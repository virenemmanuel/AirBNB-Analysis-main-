# Handle Missing Values in Numerical Columns

 Fill in missing values for numeric columns:

```python
# For columns with significant numbers of missing values, such as' weekly_price' and' monthly_price', 
# consider filling with median values or creating a separate category for missing data to avoid bias.

df['weekly_price'].fillna(df['weekly_price'].median(), inplace=True)
df['monthly_price'].fillna(df['monthly_price'].median(), inplace=True)
df['reviews_per_month'].fillna(0,inplace=True)  # Assumes missing reviews means no reviews
```

```python
# Columns like review_scores.* may contain missing values because not every listing has reviews.
# Filling with median values for review scores can be reasonable, 
# but setting missing review scores to a separate category, 
# such as 0, is also viable if it represents an absence of reviews.


# Filling missing review scores with a neutral value or 0 for absence:

df['first_review'].fillna(0, inplace=True)
df['last_review'].fillna(0, inplace=True)
```

Handeling missing values in text columns

```python
# Filling text columns with an empty string:

df['listing_url'].fillna('', inplace = True)
df['summary'].fillna('', inplace = True)
df['space'].fillna('', inplace = True)
```

```python
df.head()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number]) 

# Calculate correlation matrix
correlation_matrix = numeric_df.corr().abs()

# Identify pairs of columns with high correlation (e.g., > 0.9)
high_correlation_columns = set()
threshold = 0.9  # Adjust this threshold if needed
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if correlation_matrix.iloc[i, j] > threshold:
            colname = correlation_matrix.columns[i]
            high_correlation_columns.add(colname)

print("Highly correlated columns:\n", high_correlation_columns)
```

```python
# Count of missing values in each column
missing_counts = df.isnull().sum()
print("Missing values per column:\n", missing_counts)

# Percentage of missing values in each column
missing_percentage = (df.isnull().mean() * 100)
```
